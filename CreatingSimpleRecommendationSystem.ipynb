### Розраховуємо матрицю схожості

```python
user_movie_matrix = ratings_pd.pivot(index="userId", columns="movieId", values="rating")
```

```python
user_movie_matrix.head()
```

 Косинусна подібність

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_movie_matrix_filled = user_movie_matrix.fillna(0)
```

```python
user_movie_matrix_filled.head()
```

```python
user_similarity_cosine = cosine_similarity(user_movie_matrix_filled)
```

```python
user_similarity_df = pd.DataFrame(user_similarity_cosine, index=user_movie_matrix.index, columns=user_movie_matrix.index)
```

```python
user_similarity_df.head()
```

Прогноз оцінки

```python
import numpy as np

def predict_rating(user_id, movie_id, similarity_matrix, user_movie_matrix, k=5):
    if movie_id not in user_movie_matrix.columns:
        return np.nan  # Фільм не має оцінок
    
    users_who_rated = user_movie_matrix[movie_id].dropna().index    # Отримуємо всіх користувачів, які оцінили цей фільм
    similarities = similarity_matrix.loc[user_id, users_who_rated]  # Беремо їх схожість із нашим користувачем
    ratings = user_movie_matrix.loc[users_who_rated, movie_id]      # Беремо їхні оцінки
    top_k_users = similarities.nlargest(k).index # Відбираємо k найбільш схожих користувачів

    num = (similarities[top_k_users] * ratings[top_k_users]).sum()     # Розраховуємо зважене середнє оцінок
    den = similarities[top_k_users].abs().sum()
    
    if den == 0:
        return np.nan
    
    return num / den

predicted_rating = predict_rating(user_id=1, movie_id=2, 
                                  similarity_matrix=user_similarity_df, 
                                  user_movie_matrix=user_movie_matrix, 
                                  k=5)

print(f"Прогнозована оцінка: {predicted_rating}")
```

Розробляємо функцію рекомендацій

```python
def recommend_movies(user_id, user_movie_matrix, similarity_matrix, k=5, n_recommendations=10):
    """Функція, що генерує топ-N рекомендацій для користувача"""
    
    # Отримуємо всі фільми, які користувач ще не дивився
    unseen_movies = user_movie_matrix.columns[user_movie_matrix.loc[user_id].isna()]
    
    predictions = {}
    
    # Прогнозуємо оцінки для всіх фільмів, які користувач не бачив
    for movie_id in unseen_movies:
        predicted_rating = predict_rating(user_id, movie_id, similarity_matrix, user_movie_matrix, k)
        if not np.isnan(predicted_rating):  # Пропускаємо, якщо немає значення
            predictions[movie_id] = predicted_rating
    
    # Сортуємо за прогнозованим рейтингом та повертаємо топ-N фільмів
    top_movies = sorted(predictions.items(), key=lambda x: x[1], reverse=True)[:n_recommendations]
    
    return top_movies
```

```python
recommendations = recommend_movies(user_id=1, 
                                   user_movie_matrix=user_movie_matrix, 
                                   similarity_matrix=user_similarity_df, 
                                   k=5, n_recommendations=10)

print("Топ-10 рекомендацій (kNN):", recommendations)
```

```python
top_movies_name = []
for i in recommendations:
    movies = movies_pd[movies_pd['movieId'] == i[0]]
    top_movies_name.append(movies['title'].item())
    
print(top_movies_name)
```

Оцінка якості

```python
from sklearn.metrics import mean_squared_error

def calculate_rmse(predictions, actual):
    """Обчислення RMSE між реальними та прогнозованими оцінками"""
    mse = mean_squared_error(actual, predictions)
    return np.sqrt(mse)

# Формуємо тестовий набір (наприклад, вибираємо 20% випадкових оцінок)
test_data = ratings_pd.sample(frac=0.2, random_state=42)

predicted_ratings = [
    predict_rating(row.userId, row.movieId, user_similarity_df, user_movie_matrix, k=5) 
    for _, row in test_data.iterrows()
]

# Видаляємо NaN значення (якщо неможливо було передбачити оцінку)
valid_indices = ~np.isnan(predicted_ratings)
actual_ratings = test_data.rating[valid_indices]
predicted_ratings = np.array(predicted_ratings)[valid_indices]

rmse_knn = calculate_rmse(predicted_ratings, actual_ratings)
print(f"RMSE для kNN: {rmse_knn}")
```
